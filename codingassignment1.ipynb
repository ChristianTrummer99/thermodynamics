```
#import the proper libraries

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```
#Initialize variables
#Convert to absolute temperatures

T_L = np.arange(-40,61) 
T_H = np.arange(500,1501,10)
n_carnot = 1 - (T_L + 273)/(T_H + 273)
print(n_carnot)

#function to calculate carnot efficiency

def nCarnot(i, j):
  n_carnot = 1 - (j + 273)/(i + 273)
  value = '%.3f' % n_carnot
  return value

#print n_carnot for the given values

#a
print('\n a)', nCarnot(500,-40))
#b
print('\n b)', nCarnot(810,1))
#c
print('\n c)', nCarnot(1230, -23))
#d
print('\n d)', nCarnot(1010,11))
#e
print('\n e)', nCarnot(1500,50))
```

```
#create table with carnot efficiency values
#adjust temperature arrays

T_l = np.arange(-40,61,10) 
T_h = np.arange(500,1501,50) 

#create data array

d = []
for i in T_h:
  dataRow =[]
  for j in T_l:
    dataRow.append(nCarnot(i,j))
#     print(dataRow)
#   print('\n')
  d.append(dataRow)
# print(d)
  
table = pd.DataFrame(d, columns=T_l, index=T_h)
table




```

```
graph = plt.figure(figsize=(10, 10))

for index, row in table.iterrows():
    row_mod = np.array(row.tolist())
    plt.plot(T_l, row_mod.astype(np.float), label=str(index) + " $^\circ$C")
    
plt.legend(loc="upper right");
plt.grid(which='major', axis='both');
plt.ylabel(' n$_{carnot}$')
plt.xlabel('T$_{L}$')
```
